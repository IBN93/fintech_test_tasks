```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./NTB_Funnel.csv')
```

```python
df.head(60)
```

```python
df.status_code_1.value_counts()
```

```python
df['TSApplicationCreatedOn'] = pd.to_datetime(df['TSApplicationCreatedOn'])
df['TSApplicationModifiedOn'] = pd.to_datetime(df['TSApplicationModifiedOn'])
df['DecisionCreatedOn'] = pd.to_datetime(df['DecisionCreatedOn'])
df['DecisionModifiedOn'] = pd.to_datetime(df['DecisionModifiedOn'])
df['rep_date'] = pd.to_datetime(df['rep_date'])
```

```python
df.dtypes
```

```python
# (  3)   ,  ,   
# ( -  CR, DR, TR)
df_rates = df.groupby('rep_date', as_index=False)\
              .agg({'is_completed': sum,
                    'is_approved': sum,
                    'is_disbursed': sum,
                    'TSApplicationId': 'count'})\
              .rename(columns={'is_completed': 'is_completed_num',
                               'is_approved': 'is_approved_num',
                               'is_disbursed': 'is_disbursed_num',
                               'TSApplicationId': 'all_apps'
                              })
df_rates
```

```python
df_rates['CR'] = df_rates['is_completed_num'] / df_rates['all_apps']
df_rates['DR'] = df_rates['is_disbursed_num'] / df_rates['all_apps']
df_rates['TR'] = df_rates['is_disbursed_num'] / df_rates['is_approved_num']
```

```python
df_rates
```

```python
#     
df['steps'] = df['step_name_1'] + df['step_name_2'] + df['step_name_3']
```

```python
df_actions = df.groupby(['rep_date', 'steps'], as_index=[False, False]).agg({'steps': 'count'})\
            .rename(columns={'steps': 'number_of_times'})\
            .sort_values(['rep_date', 'number_of_times'], ascending=[False,False]).reset_index()
df_actions.head(60)
```
