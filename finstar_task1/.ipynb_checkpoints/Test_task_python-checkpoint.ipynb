```python
# В этом ноутбуке я дублирую в Python основные пункты аналитической задачи
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('./Test_sample.xlsx')
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Рассчитаем completed rate
conversion_df = df.sort_values('rep_date', ascending=True)\
                    .groupby('rep_date')\
                    .agg({'is_completed': sum,
                          'rep_date': 'count'})\
                    .rename(columns={'is_completed': 'completed_amt',
                                     'rep_date': 'all_req'})\
                    .reset_index()
conversion_df['completed_rate'] = round(conversion_df['completed_amt'] / conversion_df['all_req'], 4)
conversion_df['rep_date'] = conversion_df['rep_date'].dt.strftime('%d-%b-%y')
conversion_df
```

```python
# Построим график для Completed rate
plt.figure(figsize = (9,9))
ax = sns.lineplot(data=conversion_df, x='rep_date', y='completed_rate', marker='o', markersize=8)

ax.set_title('Completed rate')
ax.set_xlabel('Date')
ax.tick_params(axis='x', labelsize=8)
ax.set_ylabel('Completed rate')
ax.set_ylim(bottom=0, top=1)
sns.set_style("darkgrid")
```

```python
# Посмотрим статистику по страничкам, которых достиг клиент (п.6 в решении через SQL-запросы)
page_reached_df = df.sort_values('rep_date', ascending=True)\
                    .groupby(['rep_date', 'page_reached_CRM'])\
                    .agg({'is_completed': sum,
                          'rep_date': 'count'})\
                    .rename(columns={'is_completed': 'completed_amt',
                                     'rep_date': 'all_req'})\
                    .reset_index()\
                    .sort_values(['page_reached_CRM', 'rep_date'])
page_reached_df['completed_rate'] = round(page_reached_df['completed_amt'] / page_reached_df['all_req'], 4)
page_reached_df['rep_date'] = page_reached_df['rep_date'].dt.strftime('%d-%b-%y')
page_reached_df
```

```python
# Визуализируем предыдущий параграф и увидим резкий взлёт "посещаемости" страницы 
# с персональными данными (Personal Details) 20-22.05.22
plt.figure(figsize=(8,8))
ax1 = sns.lineplot(x=page_reached_df['rep_date'], 
                   y=page_reached_df['all_req'], 
                   hue=page_reached_df['page_reached_CRM'], marker='o')

ax1.set_title('Statistics of pages reached')
ax1.set_xlabel('Date')
ax1.tick_params(axis='x', labelsize=8)
ax1.set_ylabel('Number of times')
plt.legend(title='Page reached CRM')
sns.set_style("darkgrid")
```

```python
# Посмотрим статистику действий на страницах (page_status; п.8 в решении через SQL)
page_status = df.sort_values('rep_date', ascending=True)\
                    .groupby(['rep_date', 'page_reached_CRM', 'page_status'])\
                    .agg({'rep_date': 'count'})\
                    .rename(columns={'rep_date': 'all_req'})\
                    .reset_index()\
                    .sort_values(['page_reached_CRM', 'page_status', 'rep_date'])
page_status['rep_date'] = page_status['rep_date'].dt.strftime('%d-%b-%y')
page_status
```

```python
# Визуализируем предыдущий параграф и увидим, что пользователи чаще всего покидали страницу
# с персональными данными (Personal Details) 20-22.05.22. Это было самым частым событием в то время,
# как до и после сбоя самым частым событием был уход со страницы Bank Verification 
plt.figure(figsize=(8,8))
ax2 = sns.lineplot(x=page_status['rep_date'], 
                   y=page_status['all_req'], 
                   hue=page_status[['page_reached_CRM', 'page_status']].apply(tuple, axis=1), 
                   marker='o')

ax2.set_title('Statistics of actions on pages')
ax2.set_xlabel('Date')
ax2.tick_params(axis='x', labelsize=8)
ax2.set_ylabel('Number of times')
plt.legend(title='Page reached & Action on page')
sns.set_style("darkgrid")
```
